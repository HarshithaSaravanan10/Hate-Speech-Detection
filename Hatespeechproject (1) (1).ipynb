```python
dataset["labels"]=dataset["class"].map({0:"Hatespeech",1:"Offensive Language",2:"No hate or offensive language"})
```

```python
data=dataset[["tweet","labels"]]
```

```python
import re
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words("english"))
```

```python
stemmer = nltk.SnowballStemmer("english")
```

```python
def clean_data(text):
    text = str(text).lower()
    text = re.sub(r'https?://\S+|www\.S+', '', text)            # Raw string added
    text = re.sub(r'\[.*?\]', '', text)                         # Raw string added
    text = re.sub(r'<.*?>+', '', text)                          # Raw string added
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)                              # Raw string added
    text = re.sub(r'\w*\d\w*', '', text)                        # Raw string added
    text = [word for word in text.split(' ') if word not in stopwords]
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text
```

```python
data["tweet"] = data["tweet"].apply(clean_data)
```

```python
data
```

```python
x = np.array(data["tweet"])
y = np.array(data["labels"])
```

```python
x
```

```python
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv = CountVectorizer()
x = cv.fit_transform(x)
```

```python
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
X_test
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
import pickle
f = open("hate_model.pkl","wb")
pickle.dump(dt, f)
f = open("hate_vectorizer.pkl", "wb")
pickle.dump(cv, f)
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix (y_test, y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
sns.heatmap(cm, annot = True, fmt="f", cmap="YlGnBu") 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
